```python
# Netflix Stock Price Analysis üìà
# --------------------------------
# Author: Havva Zimra
# Description: A simple exploratory data analysis (EDA) and visualization of Netflix stock prices over time.
# Dataset Source: Kaggle - Netflix Stock Price History

# üß∞ Importing Required Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# üîç Exploring Available Files in the Dataset Directory
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print("Found file:", os.path.join(dirname, filename))

# üì• Load the Dataset
df = pd.read_csv("/kaggle/input/netflix-stock-price-history/Netflix_stock_data.csv")

# üïí Convert 'Date' to datetime format and sort chronologically
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

# üìä Basic Exploratory Data Analysis
print("\n--- First 5 Rows ---")
print(df.head())

print("\n--- Data Info ---")
print(df.info())

print("\n--- Statistical Summary ---")
print(df.describe())

print("\n--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Column Names ---")
print(df.columns.tolist())

# üìà Visualization: Netflix Stock Closing Price Over Time
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], color='red', linewidth=2)
plt.title('Netflix Stock Price Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Closing Price (USD)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
